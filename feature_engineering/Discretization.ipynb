```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv ('categorical.csv')
df
```

```python

# Specify the boundaries of the bins
bins_lat = [-np.inf, 0, np.inf]
# Bin labels
labels_lat = ['Latitude sul', 'Latitude norte']

# Specify the boundaries of the bins
bins_lng = [-np.inf, 0, np.inf]
# Bin labels
labels_lng = ['Longitude Oeste', 'Longitude leste']

# Bin the continuous variable ConvertedSalary using these boundaries
df['Lat'] = pd.cut(df['Latitude'], bins=bins_lat, labels=labels_lat)
df['Lng'] = pd.cut(df['Longitude'], bins=bins_lng, labels=labels_lng)

df
```
